```python
import puzzle
from collections import defaultdict

def rewrite_x(xy, regs):
  if xy[0] in regs:
    xy[0] = regs[xy[0]]
  else: xy[0] = int(xy[0])

def rewrite_y(xy, regs):
  if xy[1] in regs:
    xy[1] = regs[xy[1]]
  else: 
    xy[1] = int(xy[1])

class VM(object):
  def __init__(self, instrs):
    self.pc = 0
    self.regs = defaultdict(int)
    self.regs['a'] = 0
    self.regs['b'] = 0
    self.regs['c'] = 0
    self.regs['d'] = 0
    self.regs['e'] = 0
    self.regs['f'] = 0
    self.regs['g'] = 0
    self.regs['h'] = 0
    self.instrs = instrs

  def step(self):
    self.blocking = False
    self.sent = False
    inst = self.instrs[self.pc]
    op, xy = inst.split()[0], inst.split()[1:]
    # print(op, xy)
    if op in ['jnz']:
      rewrite_x(xy, self.regs)
    if op in ['add', 'mul', 'sub', 'jnz', 'set']:
      rewrite_y(xy, self.regs)
    if op in ['add', 'mul', 'sub', 'jnz', 'set']:
      x, y = xy
    else:
      x = xy[0]

    if op == 'set':
      self.regs[x] = y
    elif op == 'add':
      new = self.regs[x] + y
      self.regs[x] = new
    elif op == 'mul':
      new = self.regs[x] * y
      self.regs[x] = new
      self.regs['mul'] += 1
    elif op == 'sub':
      new = self.regs[x] - y
      self.regs[x] = new
    elif op == 'jnz':
      if x != 0:
        # cancel out the += 1 at the end of the loop
        self.pc += y -1
    self.pc += 1

```

```python
def two(INPUT):
  vm = VM(instrs=INPUT)
  vm.regs['a'] = 1
  while True:
    vm.step()
    if vm.pc == 26:
      print(vm.regs)
  return 0

p = puzzle.Puzzle("", "23")
p.run(two, 0)
```
